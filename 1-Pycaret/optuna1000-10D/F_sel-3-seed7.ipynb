```python
import numpy as np
import pandas as pd
df = pd.read_csv('descriptors-alva-66-fix.csv')
ncol = df.shape[1]-1
x = df.iloc[:,0:ncol]
y = df['active']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 7)
```

```python
trainD = pd.concat([x_train, y_train], axis =1)
testD = pd.concat([x_test, y_test], axis =1)
```

```python
F_sel = ['O%',
 'ZM1Kup',
 'MAXDN',
 'EE_B(s)',
 'Eig08_AEA(dm)',
 'Mor24s',
 'HATS6u',
 'CATS2D_03_AL',
 'TPSA(NO)',
 'MLOGP'
]
trainD = trainD[['O%',
 'ZM1Kup',
 'MAXDN',
 'EE_B(s)',
 'Eig08_AEA(dm)',
 'Mor24s',
 'HATS6u',
 'CATS2D_03_AL',
 'TPSA(NO)',
 'MLOGP','active']]
testD = testD[['O%',
 'ZM1Kup',
 'MAXDN',
 'EE_B(s)',
 'Eig08_AEA(dm)',
 'Mor24s',
 'HATS6u',
 'CATS2D_03_AL',
 'TPSA(NO)',
 'MLOGP','active']]
```

```python
from pycaret.classification import *
cm = setup(data = trainD, target = 'active', train_size = 0.8, test_data = testD, silent=True,  numeric_features = F_sel,
             session_id=7, fold_strategy = 'stratifiedkfold', fold = 5, fold_shuffle = True, normalize = True)
```

```python
compare_models()
```
